```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import tensorflow as tf
import tensorflow_hub as hb
from tensorflow import keras
import skimage
import sklearn
import cv2
from google.colab import drive
import random
import os
%matplotlib inline
```

```python
drive.mount('/content/drive')
```

```python
load_inception_model = tf.keras.models.load_model("/content/drive/MyDrive/Intel_Image_classification/models_inception_intel_image_model.h5",custom_objects={'KerasLayer':hb.KerasLayer})
```

```python
load_inception_model.compile(loss="CategoricalCrossentropy()",optimizer="adam",metrics=["accuracy"])
```

```python
names_labels = ['mountain','street','glacier','buildings','sea','forest']
```

### Test Case 1:-

```python
image1 = "/content/drive/MyDrive/Intel_Image_classification/seg_pred/61.jpg"
image = mpimg.imread(image1)
pres =load_inception_model.predict(np.array([image]))
name = names_labels[np.argmax(pres)]
print(name)
plt.imshow(image)
plt.axis('off')
plt.title(name)
plt.show()
```
